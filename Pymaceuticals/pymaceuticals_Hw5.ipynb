```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress


# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
#Merge the csv files to create a dataframe
mouse_study = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])
mouse_study
```

```python
#Calculate the number of unique mice
mouse_number = mouse_study["Mouse ID"].nunique()
mouse_number
```

```python
#Locate the duplicate mice in the merged dataframe 
Dup_mouse_ID = mouse_study.loc[mouse_study.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
#Dup_mouse_ID[Dup_mouse_ID["Mouse ID".isin(ids[ids.duplicated()])].sort_values("ID")
print(Dup_mouse_ID)
```

```python
#Dataframe for each the following non-duplicate mice
mouse_clean_study = mouse_study.drop_duplicates(subset=['Mouse ID'])
mouse_clean_study
```

```python
#Number of mice in the clean dataframe 
mouse_number_clean = mouse_clean_study["Mouse ID"].nunique()
mouse_number_clean 
```

```python
#Summary statistics for the original merged dataframe 
mean_mouse = mouse_study["Tumor Volume (mm3)"].mean()
print(f"The mean of the tumor sizes {mean_mouse}.")
median_mouse = mouse_study["Tumor Volume (mm3)"].median()
print(f"The median of tumor sizes {median_mouse}.")
var_mouse = np.var(mouse_study["Tumor Volume (mm3)"],ddof = 0)
print(f"The tumor size variance {var_mouse}")
sd_mouse = np.std(mouse_study["Tumor Volume (mm3)"],ddof = 0)
print(f"The standard deviation of the tumor sizes {sd_mouse}")
#Error_mouse = st(mouse_study["Tumor Volume (mm3)"])
sem_mouse = mouse_study["Tumor Volume (mm3)"].sem(axis = 0)
#sems = mouse_study.groupby("Tumor Volume (mm3)").sem()
print(f"The standard error of the mean of the tumor sizes is {sem_mouse}")
```

```python
#Creates of dataframe for the summary statistics.
Summary_mouse_study = pd.DataFrame({"Average Tumor size": [mean_mouse],
                                   "Median Tumor size": [median_mouse],
                                   "Variance of Tumor size": [var_mouse],
                                   "Standard Devitation of the Tumor Size": [sd_mouse],
                                   "Standard Error of Tumor Sizes": [sem_mouse]})
Summary_mouse_study
```

```python
#Create summary statistics for the clean dataframe
Summary_mouse_study_clean = mouse_clean_study.aggregate({"Tumor Volume (mm3)":['mean', 'median', 'var', 'std', 'sem']})
Summary_mouse_study_clean
#grouped_single = df.groupby('Team').agg({'Age': ['mean', 'min', 'max']})
```

```python
quartiles1 = Tumor_Capo_Result.quantile([.25,.5,.75])
quartiles2 = Tumor_Rami_Result.quantile([.25,.5,.75])
quartiles3 = Tumor_Infu_Result.quantile([.25,.5,.75])
quartiles4 = Tumor_Ceft_Result.quantile([.25,.5,.75])
lowerq1 = quartiles1[0.25]
upperq1 = quartiles1[0.75]
lowerq2 = quartiles2[0.25]
upperq2 = quartiles2[0.75]
lowerq3 = quartiles3[0.25]
upperq3 = quartiles3[0.75]
lowerq4 = quartiles4[0.25]
upperq4 = quartiles4[0.75]
iqr1 = upperq1-lowerq1
iqr2 = upperq2-lowerq2
iqr3 = upperq3-lowerq3
iqr4 = upperq4-lowerq4
lower_bound1 = lowerq1 - (1.5*iqr1)
upper_bound1 = upperq1 + (1.5*iqr1)
lower_bound2 = lowerq2 - (1.5*iqr2)
upper_bound2 = upperq2 + (1.5*iqr2)
lower_bound3 = lowerq3 - (1.5*iqr3)
upper_bound3 = upperq3 + (1.5*iqr3)
lower_bound4 = lowerq4 - (1.5*iqr4)
upper_bound4 = upperq4 + (1.5*iqr4)
print(f"Values of Capomulin below {lower_bound1} could be outliers.")
print(f"Values of Capomulin above {upper_bound1} could be outliers.")
print(f"Values of Ramicane below {lower_bound2} could be outliers.")
print(f"Values of Ramicane above {upper_bound2} could be outliers.")
print(f"Values of Infubinol below {lower_bound3} could be outliers.")
print(f"Values of Infubinol above {upper_bound3} could be outliers.")
print(f"Values of Ceftamin below {lower_bound4} could be outliers.")
print(f"Values of Ceftamin above {upper_bound4} could be outliers.")
```

```python
#Create a scatterplot between the size of the tumor and the weight of the mice for the drug Capomulin
Capo_weight = Capomulin_total_drug['Weight (g)']
Capo_tumor = Capomulin_total_drug['Tumor Volume (mm3)']
plt.scatter(Capo_weight, Capo_tumor, marker="o", facecolors="red", edgecolors="black")
plt.title("Capomulin Drug Results")
plt.xlabel("Weight")
plt.ylabel("Tumor Size")
```

```python
#Caculates the correlation coefficient for the Capomulin scatterplot
print(f"The correlation coefficient between the tumor size and the weight of the mouse is {round(st.pearsonr(Capo_weight,Capo_tumor)[0],2)}")
```

```python
#Calculates the trend line for the Capomulin scatterplot
Capo_weight = Capomulin_total_drug['Weight (g)']
Capo_tumor = Capomulin_total_drug['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(Capo_weight, Capo_tumor)
regress_values = Capo_weight *slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(Capo_weight, Capo_tumor, marker="o", facecolors="red", edgecolors="black")
plt.plot(Capo_weight,regress_values,"b-")

plt.title("Capomulin Drug Results")
plt.xlabel("Weight")
plt.ylabel("Tumor Size")
```
